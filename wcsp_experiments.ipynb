```python
import os
import json
import subprocess
from pathlib import Path
```

```python
input_dir = "donnees/donnees_wcsp"
output_dir = "src/vcsp/instances"
max_cost = 1000
max_time = 600
```

**max_cost = 1000;      contraintes dures = max_cost;      contraintes_souples = 1**

```python
os.makedirs(output_dir, exist_ok=True)
solution_dir = os.path.join(output_dir, "solutions")
os.makedirs(solution_dir, exist_ok=True)

for file_name in os.listdir(input_dir):
    if file_name.endswith(".json"):
        json_path = os.path.join(input_dir, file_name)
        print(f"JSON Path : {json_path}")

        wcsp_file_name = Path(file_name).stem + ".wcsp"
        wcsp_path = os.path.join(output_dir, wcsp_file_name)
        print(f"Traitement de {wcsp_file_name}...")

        command = [
                "python", "src/vcsp/wcsp.py",
                json_path, wcsp_path,
                "--max_cost", "1000"
            ]

        try:
            result = subprocess.run(command, check=True, capture_output=True, text=True)
            print(result.stdout)
        except subprocess.CalledProcessError as e:
            print(f"Erreur lors du traitement de {wcsp_file_name} :")
            print(e.stderr)
        print('\n')
```
